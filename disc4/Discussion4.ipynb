```python
"""
Categorical data
1) Bar plots -> good for comparing categorical data
2) Count plot -> shows frequency of occurences

Continuous data
1) Histogram -> distribution of numerical data
    - Histogram vs. count plot
    - Count plot displays frequencies of each unique category
    - Histogram displays the distributions of numerical data by grouping data into bins/intervals
2) KDE Plot -> probability density function of continuous data
    - PDF shows the likelihood of a continuous random variable taking a variable value

Bivariate analysis (2 variables)
1) Scatter plot -> shows relationships between two continuous variables
2) Pair plot -> shows all relationships between all possibilities of variables (compares 2 at a time)

Multivariate analysis (multiple variables/dimensions)
1) Heatmap -> visualizes relationship between variables using color coded matrix)
2) Box plot -> shows distribution of data and can be used to detect outliers

Regression plot -> shows relationship and line between two variables

"""
```

```python
# Q7: Plot a bar chart showing the average tip amount by day using the tips dataset

tips = sns.load_dataset("tips")
# tips
by_day = tips.groupby("day").mean("tip")
# by_day
sns.barplot(by_day, x = "day", y = "tip")
plt.title("Average tip amount by day")
```
